```python
age = np.array(DF.age)
sym = ["weeks", "+", "-", " "]
for x,y in enumerate(age):
    if type(y) is float:
        age[x] = y
    elif sym[0] in y:
        age[x] = 0
    elif sym[1] in y:
        age[x] = y.replace('+','')
        age[x] = int(round(float(age[x])))
    elif sym[2] in y:
        if len(y) >= 4:
            mat = y.split("-")
            if "Sep" in mat:
                age[x] = 0
            else:
                age[x] = (int(mat[0]) + int(mat[1]))/2
        else:
            age[x] = y.replace('-','')
            age[x] = int(round(float(age[x])))
    elif sym[3] in y:
        num, month = y.split(" ")
        num = int(num)
        if num < 6:
            age[x] = 0
        elif num >= 6 and num < 18:
            age[x] = 1
        else:
            age[x] = 2
for x,y in enumerate(age):
    age[x] = float(y)
DF["age"] = age
DF["age"].fillna(DF["age"].mean(), inplace = True)
```
